```python
##Import your libraries: Beautiful soup, urllib, and re (For regular expressions)

from urllib.request import urlopen
from bs4 import BeautifulSoup
import re
```

```python
# read the URL, and put the HTML page into beautiful soup

raw_html = urlopen("http://floatingmedia.com/columbia/BBC.html").read()
soup_doc = BeautifulSoup(raw_html, "html.parser")
```

```python
#find_all
all_divs = soup_doc.find(class_='body-content')
all_p = all_divs.find_all('p')
all_p
```

```python
##TakeYou're working loop And add the find_all for each_movie
#And the inner loop that loops through each_movie

movie_list = []
for p_line in all_p[:-4]:
    if p_line.strong is not None:
        #print(critic_info)
        critic_info = p_line.strong.string
        #print(critic_info)
        movie_info = p_line.next_sibling
        #print(movie_info)
        regex_for_name = re.findall(r"^[A-Z][^–]+", critic_info)
        regex_for_org = re.findall(r"– (.*) [(]", critic_info) 
        regex_for_cn = re.findall(r"\(.*\)", critic_info)
        this_ten_movies = movie_info.find_all(string = True)
        #print(this_ten_movies)
        for each_movie in this_ten_movies:
            #print(each_movie, critic_info)
            #print(critic_info)
            this_movie = [] # ----------------> put in critic information and put in movie information
            regex_for_mname = r"\d.(.*)\("
            regex_for_year = r"(\d{4})\)$"
            regex_for_director = r"([\w\s-]+), \d{4}\)$"
            movie_director = re.findall(regex_for_director,each_movie)
            movie_year = re.findall(regex_for_year,each_movie)
            movie_mname = re.findall(regex_for_mname,each_movie)
            this_movie.append(movie_mname[0])
            this_movie.append(movie_year[0])
            this_movie.append(movie_director[0])
            this_movie.append(regex_for_name[0])
            this_movie.append(regex_for_org[0])
            this_movie.append(regex_for_cn[0])
            
            
            movie_list.append(this_movie)
            
        #print(critic_info, movie_info)
```
